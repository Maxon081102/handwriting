```python
def new_photo(name):
    sample = Image.open('data/Lectures1-6/' + name)
    s = []
    for i in range(0, 2479):
        for j in range(0, 3504):
            a, b, c = sample.getpixel((i, j))
            if a > 180 or b > 180 or c > 180:
                sample.putpixel((i , j), (255, 255 , 255))
            if a < 80 and b < 80 and c < 80:
                s.append([i, j])
    sample2 = sample.resize((768, 1024))
    s = []
    for i in range(0, 768):
        for j in range(0, 1024):
            a, b, c = sample2.getpixel((i, j))
            if a < 250 and b < 250 and c < 250:
                s.append([i, j])
    for elem in s:
        sample2.putpixel((elem[0] , elem[1]), (120, 120, 120))
    sample2.save('data/PLectures1-6/' + name)
    
    
```

```python
for elem in catalog:
    if 'jpg' in elem:
        new_photo(elem)
```
